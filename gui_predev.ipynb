```python
from tkinter import *
import customtkinter as tk
from PIL import Image
import csv
#from tkVideoPlayer import TkinterVideo

#setting the appearance of the login window, built in apperance mode "dark"
tk.set_appearance_mode("dark")
tk.set_default_color_theme("blue")
titan_app = tk.CTkToplevel()

#create the geometry and the window name for the app
titan_app.title("---Titan App---")
titan_app.geometry("650x650")
#titan_app.config(bg="#7b337d")
titan_app.wm_attributes("-fullscreen",False)
titan_app.state=("normal")

backround_img = tk.CTkImage(
    light_image=Image.open("image_files/wobbly_image.png"),
    dark_image=Image.open("image_files/wobbly_image.png"),
    size=(2000,900)
    )

image_label = tk.CTkLabel(master=titan_app, 
    text="",
    width=650,
    height=650,
    image=backround_img
    )
image_label.place(relx=0.5, rely=0.5, relwidth=1.0, relheight=1.0, anchor="center")


login_frame = tk.CTkFrame (master=titan_app,
    fg_color="white",
    #fg_color="#ff65f3",
    bg_color="transparent",
    width=400,
    height=400,
    corner_radius=15,
    border_width=1,
    border_color="grey",
    background_corner_colors=("black","black","black","black",)
    #background_corner_colors=("white","white","white","white",)
    )
login_frame.place (relx=0.5, rely=0.5, anchor="center")


username_entry = tk.CTkEntry(master=login_frame,
    width=75,
    height=40,
    placeholder_text="Username",
    placeholder_text_color="white",
    corner_radius=15,
    font=(("comicsans",20)),
    fg_color="black",
    border_width=0.95,
    border_color="white",
    text_color="white"
    )
username_entry.place(relx=0.5, rely=0.37, relwidth=0.97, relheight=0.23, anchor="s")

password_entry = tk.CTkEntry(master=login_frame,
    width=75,
    height=40,
    placeholder_text="Password",
    placeholder_text_color="white",
    corner_radius=15,
    show="*",
    font=(("comicsans",20)),
    fg_color="black",
    #border_width=0.95,
    #border_color="white",
    text_color="white"
    )
password_entry.place(relx=0.5, rely=0.61, relwidth=0.97, relheight=0.23, anchor="s")


def login_check():
    with open("login_info.csv", "rt", ) as csv_file:
        reader = csv.reader(csv_file, delimiter=",")
        found = False
    for row in reader:
        if username_entry == row[0] and password_entry == row[1]:
            found = True
            username_entry.configure(text="access granted")
    if found == False:
            username_entry.configure(text="access denied")
            
#log in button, command will be to check login details
log_button = tk.CTkButton(master=login_frame, 
    text="Log In",
    font=(("comicsans",20)),
    text_color="white",
    fg_color="#343638",
    #fg_color="#d622dc",
    hover_color="black",
    corner_radius=10,
    height=75,
    width=40,
    #border_width=0.5,
    #border_color="white",
    command=login_check
    )
log_button.place(relx=0.505, rely=0.715, relwidth=0.48, anchor="w")
    
#sign up button, command will be to create account details in database
signup_button = tk.CTkButton(master=login_frame, 
    text="Sign up",
    font=(("comicsans",20)),
    text_color="white",
    fg_color="#343638",
    #fg_color="#d622dc",
    hover_color="black",
    corner_radius=10,
    height=75,
    width=40,
    #border_width=0.5,
    #border_color="white",
    )
signup_button.place(relx=0.495, rely=0.715, relwidth=0.48, anchor="e")

def dark_mode():
    tk.set_appearance_mode("dark")
    
#button to enable user to change password when it forgotten
forgot_pass_button = tk.CTkButton(master=login_frame, 
    text="forgot password",
    font=(("comicsans",20)),
    text_color="white",
    fg_color="#343638",
    #fg_color="#d622dc",
    hover_color="black",
    corner_radius=10,
    height=65,
    width=45,
    #border_width=0.5,
    #border_color="white"
    command=dark_mode
    )
forgot_pass_button.place(relx=0.5, rely=0.82, relwidth=0.97, anchor="n")
#relwidth=0.981

app_label = tk.CTkLabel(master=titan_app,
    bg_color="white",
    text="TITAN",
    text_color="white",
    font=(("comicsans",40)),
    width=100,
    height=60,
    )
app_label.place(relx=0.5, rely=0.065, relwidth=1.0, anchor="s")

app_label_2 = tk.CTkLabel(master=titan_app,
    bg_color="white",
    text="",
    text_color="white",
    font=(("comicsans",40)),
    width=100,
    height=60,
    )
app_label_2.place(relx=0.5, rely=1.0, relwidth=1.0, anchor="s")


titan_app.mainloop()
```
