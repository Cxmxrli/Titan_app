```python
import customtkinter as tk
from PIL import Image


#setting the appearance of the login window, built in apperance mode "dark"
tk.set_appearance_mode("dark")
tk.set_default_color_theme("blue")
titan_app = tk.CTk()

#create the geometry and the window name for the app
titan_app.title("---Titan App---")
titan_app.geometry("700x700")

#CTkImage defines the images for light and dark mode it uses Image imported from PIL
backround_img = tk.CTkImage(
    dark_image=Image.open("image_files/wobbly_image.png"),
    light_image=Image.open("image_files/light_image.png"),
    size=(2000,1000)
)

#actual label where the image is placed, the size of this determines the size of the image or how it appears
image_label = tk.CTkLabel(master=titan_app, 
    text="",
    width=650,
    height=650,
    image=backround_img
)
image_label.place(relx=0.5, rely=0.5, relwidth=1.0, relheight=1.0, anchor="center")

#frame wich contains all elements that the user interacts with
login_frame = tk.CTkFrame(master=titan_app,
    fg_color="black",
    width=400,
    height=400,
    corner_radius=30,
    border_width=0.9,
    border_color="white",
    background_corner_colors=("black,") ,
    image = backround_img
)
login_frame.place(relx=0.5, rely=0.5, anchor="center")

info_text = tk.CTkLabel(master=login_frame,
    fg_color="black",
    width=75,
    height=5,
    text="Enter username & password",
    text_color="white",
    font=(("comicsans",20)),        
)
info_text.place(relx=0.5, rely=0.125, relwidth=0.7, relheight=0.10, anchor="s")


#entry box for entering username and also entry box for entering password
#will be checked against database
username_entry = tk.CTkEntry(master=login_frame,
    width=75,
    height=40,
    placeholder_text="Username",
    placeholder_text_color="grey",
    corner_radius=25,
    font=(("comicsans",20)),
    fg_color="black",
    border_width=0.95,
    border_color="white",
    text_color="white"
)
username_entry.place(relx=0.5, rely=0.37, relwidth=0.97, relheight=0.23, anchor="s")

password_entry = tk.CTkEntry(master=login_frame,
    width=75,
    height=40,
    placeholder_text="Password",
    placeholder_text_color="grey",
    corner_radius=25,
    show="*",
    font=(("comicsans",20)),
    fg_color="black",
    border_width=0.95,
    border_color="white",
    text_color="white"
)
password_entry.place(relx=0.5, rely=0.61, relwidth=0.97, relheight=0.23, anchor="s")

#button for login, command will later be assigned to checking the input of a text box
#against database
log_button = tk.CTkButton(master=login_frame, 
    text="Log In",
    font=(("comicsans",20)),
    text_color="white",
    fg_color="#343638",
    hover_color="black",
    corner_radius=30,
    height=75,
    width=40,
)
log_button.place(relx=0.495, rely=0.715, relwidth=0.48, anchor="e")

#another button, this one foe sign up, will be used to prompt the user to make an account
signup_button = tk.CTkButton(master=login_frame, 
    text="Sign up",
    font=(("comicsans",20)),
    text_color="white",
    fg_color="#343638",
    hover_color="black",
    corner_radius=30,
    height=75,
    width=40,
)
signup_button.place(relx=0.505, rely=0.715, relwidth=0.48, anchor="w")

#button for when a password is forgotten, prompts the user to enter a passkey to change password
forgot_pass_button = tk.CTkButton(master=login_frame, 
    text="forgot password",
    font=(("comicsans",20)),
    text_color="white",
    fg_color="#343638",
    hover_color="black",
    corner_radius=30,
    height=65,
    width=45,
)
forgot_pass_button.place(relx=0.5, rely=0.82, relwidth=0.97, anchor="n")

titan_app.mainloop()
```
